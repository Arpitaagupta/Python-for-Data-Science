# What is Dataframe and its operations?

A dataframe is similar to Excel workbook tabular data structure.

### 1. DataFrame.shape() : 
Dimensions of data frames " Rows x Columns " 

### 2. DataFrame.head(n) / DataFrame.tail(n) :
Access top / bottom n records

### 3. DataFrame.columns
Access all columns

### 4. DataFrame["column_name"]
Access data of one column

### 5. DataFrame[["column1", "column2",...]]
Access data of multiple columns
